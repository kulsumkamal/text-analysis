```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
```

```python
col_names = ['x','ids', 'date', 'QUERY', 'Username', 'Statement']
data = pd.read_csv('../Downloads/training.csv', encoding = 'latin1', names = col_names, header=None)
```

```python
data.shape
```

```python
data.head()
```

```python
data = data.drop(['x'], axis=1)
data.head()
```

```python
data.info()
```

```python
labels = []

for x,y in data.Statement.iteritems():
    label = TextBlob(y)
    labels.append(label.sentiment.polarity)

labels
```

```python
data['label'] = labels
```

```python
data.head()
```

```python
def pol_to_label(x):
    if x>0 and x<=1:
        return 'pos'
    elif x==0:
        return 'neutral'
    elif x<0 and x>=-1:
        return 'neg'
    pass

data.label = data.label.apply(pol_to_label)
```
