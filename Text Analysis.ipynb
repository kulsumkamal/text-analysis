```python
import requests
from bs4 import BeautifulSoup as soup
from urllib.request import Request as uReq
from urllib.request import urlopen
import pandas as pd
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36'
headers = {'User-Agent':user_agent}
```

```python
input_file = pd.read_excel("../Downloads/shmt/Input.xlsx")
```

```python
input_file.head()
```

```python
page = []
for url in input_file.URL:
    req = uReq(url, None, headers)
    response = urlopen(req)
    page_html = response.read()
    page_soup = soup(page_html, 'html.parser')
    page.append(page_soup)
```

```python
page[0].findAll('p')
```

```python
page[0].findAll('title')
```

```python
titles = []

for page_soup in page:
    titles.append(' '.join(list(map(lambda x:str(x).replace('<title>','').replace('</title>',''),page_soup.findAll('title')))))
```

```python
titles
```

```python
content = []

for page_soup in page:
    content.append(' '.join(list(map(lambda x:str(x).replace('<p>','').replace('</p>',''),page_soup.findAll('p')))))
```

```python
content[0]
```

```python
input_file['content']=content
```

```python
input_file['titles']=titles
```

```python
input_file.head()
```

```python
pronouns = ['i', 'he', 'him', 'her', 'it', 'me', 'she', 'them', 'they', 'us', 'we', 'you']
```

```python
import re
```

```python
dictionary = pd.read_csv("C:/Users/kulsum kamal/Downloads/LoughranMcDonald_MasterDictionary_2020.csv")
```

```python
dictionary.head()
```

```python
stopwords = open("C:/Users/kulsum kamal/Downloads/StopWords_Generic.txt",'r')
stopwords = stopwords.read().split()
```

```python
avg_word_count = []
avg_word_len = []
word_count = []
syll_count = []
complex_count = []
complex_percent = []
fog_index = []
pos_score = []
neg_score = []
pol_score = []
sub_score = []
pronoun_count = []

for article in content:
    sentences = article.split('.')
    num_sen = len(sentences)
    num_words = 0
    num_syll = 0
    num_complex = 0
    pos_num = 0
    neg_num = 0
    num_pro = 0
    num_char = 0
    for text in sentences:
        text = re.sub(r'\d+(?:\.\d*)?(?:[eE][+-]?\d+)?', '', text)
        text = re.sub(r'\W+', ' ', text, flags=re.M)
        for word in text.split():
            if word.lower() in pronouns and word!='US':
                num_pro = num_pro + 1
            if not word in stopwords:
                f = word.upper() in dictionary['Word'].tolist()
                num_char = num_char + len(word)
                num_syll = num_syll + dictionary[dictionary['Word']==word.upper()].Syllables.item() if f else 0
                if f and dictionary[dictionary['Word']==word.upper()].Syllables.item()>2:
                    num_complex = num_complex + 1
                if f and dictionary[dictionary['Word']==word.upper()].Positive.item()!=0:
                    pos_num = pos_num + 1
                if f and dictionary[dictionary['Word']==word.upper()].Negative.item()!=0:
                    neg_num = neg_num + 1
                num_words = num_words + 1
    
    pronoun_count.append(num_pro)
    sub_score.append((pos_num+neg_num)/(num_words+0.000001))
    pol_score.append((pos_num-neg_num)/(pos_num+neg_num+0.000001))
    neg_score.append(neg_num)
    pos_score.append(pos_num)
    fog_index.append(0.4*((num_words/num_sen)+(num_complex/num_words)))
    complex_percent.append(num_complex/num_words)
    complex_count.append(num_complex)
    syll_count.append(num_syll/num_words)
    avg_word_count.append(num_words//num_sen)
    avg_word_len.append(num_char//num_words)
    word_count.append(num_words)
```
